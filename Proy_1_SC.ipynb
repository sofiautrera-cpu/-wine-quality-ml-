### Modelos de Clasificación: Random Forest y XGBoost

Dado que hemos cambiado la variable objetivo a `primary_category`, ahora abordaremos un problema de clasificación. Entrenaremos y evaluaremos modelos `RandomForestClassifier` y `XGBClassifier`.

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Identify categorical and numerical features for the *new* X
# Drop specific columns that are not suitable for direct modeling or have too many missing values/high cardinality
cols_to_drop = [
    'Unnamed: 0', 'product_id', 'product_name', 'variation_desc',
    'ingredients', 'highlights', 'price_usd' # price_usd is now a feature, but might be less relevant for category prediction
]

# Filter columns that actually exist in X_train and X_test
existing_cols_to_drop_train = [col for col in cols_to_drop if col in X_train.columns]
existing_cols_to_drop_test = [col for col in cols_to_drop if col in X_test.columns]

X_train_filtered_clf = X_train.drop(columns=existing_cols_to_drop_train, errors='ignore')
X_test_filtered_clf = X_test.drop(columns=existing_cols_to_drop_test, errors='ignore')

numerical_features_clf = X_train_filtered_clf.select_dtypes(include=['int64', 'float64']).columns
categorical_features_clf = X_train_filtered_clf.select_dtypes(include=['object']).columns

# Create preprocessing pipelines for numerical and categorical features
numerical_transformer_clf = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))
])

categorical_transformer_clf = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Create a column transformer to apply different transformations to different columns
preprocessor_clf = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer_clf, numerical_features_clf),
        ('cat', categorical_transformer_clf, categorical_features_clf)
    ])

# --- Random Forest Classifier ---

# Create the Random Forest Classifier model pipeline
model_rf_clf = Pipeline(steps=[('preprocessor', preprocessor_clf),
                               ('classifier', RandomForestClassifier(random_state=42, n_estimators=100))])

# Train the model
model_rf_clf.fit(X_train_filtered_clf, y_train)

# Make predictions
y_pred_rf_clf = model_rf_clf.predict(X_test_filtered_clf)

# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf_clf)
precision_rf = precision_score(y_test, y_pred_rf_clf, average='weighted', zero_division=0)
recall_rf = recall_score(y_test, y_pred_rf_clf, average='weighted', zero_division=0)
f1_rf = f1_score(y_test, y_pred_rf_clf, average='weighted', zero_division=0)

print("Resultados del modelo Random Forest Classifier:")
print(f"Accuracy: {accuracy_rf:.2f}")
print(f"Precision (weighted): {precision_rf:.2f}")
print(f"Recall (weighted): {recall_rf:.2f}")
print(f"F1-Score (weighted): {f1_rf:.2f}")

# --- XGBoost Classifier ---

# Create the XGBoost Classifier model pipeline
# XGBoost requires integer labels for classification if the objective is 'multi:softmax' or 'multi:softprob'
# We'll need to encode y_train and y_test if they are string labels.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

model_xgb_clf = Pipeline(steps=[('preprocessor', preprocessor_clf),
                                ('classifier', XGBClassifier(objective='multi:softmax', random_state=42, n_estimators=100, use_label_encoder=False, eval_metric='mlogloss'))])

# Train the XGBoost model
model_xgb_clf.fit(X_train_filtered_clf, y_train_encoded)

# Make predictions with XGBoost
y_pred_xgb_clf_encoded = model_xgb_clf.predict(X_test_filtered_clf)
y_pred_xgb_clf = le.inverse_transform(y_pred_xgb_clf_encoded)

# Evaluate the XGBoost model
accuracy_xgb = accuracy_score(y_test, y_pred_xgb_clf)
precision_xgb = precision_score(y_test, y_pred_xgb_clf, average='weighted', zero_division=0)
recall_xgb = recall_score(y_test, y_pred_xgb_clf, average='weighted', zero_division=0)
f1_xgb = f1_score(y_test, y_pred_xgb_clf, average='weighted', zero_division=0)

print("\nResultados del modelo XGBoost Classifier:")
print(f"Accuracy: {accuracy_xgb:.2f}")
print(f"Precision (weighted): {precision_xgb:.2f}")
print(f"Recall (weighted): {recall_xgb:.2f}")
print(f"F1-Score (weighted): {f1_xgb:.2f}")
```
